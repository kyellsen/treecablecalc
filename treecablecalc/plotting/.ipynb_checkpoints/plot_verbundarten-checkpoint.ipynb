```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_simple_connection():
    G = nx.Graph()
    G.add_edge(0, 1)
    pos = {0: (0, 0), 1: (1, 0)}
    nx.draw(G, pos, with_labels=False, node_size=3000, node_color='lightblue')
    plt.title("Einfach-Verbindung")
    plt.show()
    
draw_simple_connection()
```

```python
def draw_triangle_connection():
    G = nx.Graph()
    G.add_edges_from([(0, 1), (1, 2), (2, 0)])
    pos = {0: (0, 1), 1: (1, 0), 2: (0, 0)}
    nx.draw(G, pos, with_labels=False, node_size=3000, node_color='lightblue')
    plt.title("Dreiecks-Verbindung")
    plt.show()
    
draw_triangle_connection()
```

```python
def draw_multiple_triangle_connection():
    G = nx.Graph()
    G.add_edges_from([(0, 1), (1, 2), (2, 0), (2, 3), (3, 0)])
    pos = {0: (0, 1), 1: (1, 0), 2: (0, 0), 3: (-1, 0)}
    nx.draw(G, pos, with_labels=False, node_size=3000, node_color='lightblue')
    plt.title("Mehrfach-Verbindung")
    plt.show()
    
draw_multiple_triangle_connection()
```

```python
def draw_ring_connection():
    G = nx.Graph()
    G.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 0)])
    pos = {0: (0, 1), 1: (1, 1), 2: (1, 0), 3: (0, 0)}
    nx.draw(G, pos, with_labels=False, node_size=3000, node_color='lightblue')
    plt.title("Ring-Verbindung")
    plt.show()



# Draw all connection type

draw_ring_connection()
```

```python
def draw_central_connection():
    G = nx.Graph()
    central_node = 5
    outer_nodes = [0, 1, 2, 3, 4]
    
    # Kanten hinzufgen
    G.add_edges_from([(node, central_node) for node in outer_nodes])
    
    # Anzahl der ueren Knoten
    num_outer_nodes = len(outer_nodes)
    
    # Berechnung der Positionen der ueren Knoten auf einem Kreis
    angle_step = 2 * np.pi / num_outer_nodes
    pos = {i: (np.cos(i * angle_step), np.sin(i * angle_step)) for i in outer_nodes}
    
    # Position des zentralen Knotens
    pos[central_node] = (0, 0)
    
    # Zeichnen des Graphen
    plt.figure(figsize=(8, 8))
    nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=12)
    plt.title("Zentral-Verbindung")
    plt.show()

draw_central_connection()
```
